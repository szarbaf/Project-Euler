```python
from collections import Counter
import math
import itertools

def get_digits(d):
  out = []
  while d != 0:
    out.append(d % 10)
    d //= 10
  out.reverse()
  return out
# def is_permute
def to_number(d):
  out = 0
  for i in d:
    out = 10*out + i
  return(out)

def is_prime(d):
  counter = 2
  while counter <= math.sqrt(d):
    if d % counter == 0:
      return False
    counter += 1

  return True


def is_permutation(a, b):
  a_count = Counter(a)
  b_count = Counter(b)

  # import pdb
  # pdb.set_trace()
  if len(a_count) != len(b_count):
    return False

  for e, c in a_count.items():
    # print(e, c, b_count[e])
    if b_count[e] != c :
      return False

  return True

def get_permutations(d):
  return(itertools.permutations(d))
```

```python
counter = 1000

final_out = set()
while counter <= 9999:
  cur_perm = set(get_permutations(get_digits(counter)))

  if len(cur_perm) >= 3:
    cur_perm = list(cur_perm)

    for i in range(len(cur_perm) - 2):
      for j in range(i+1, len(cur_perm) - 1):
        for k in range(j+1, len(cur_perm)):

          cur = [to_number(el) for el in [cur_perm[i], cur_perm[j], cur_perm[k]]]
          cur.sort()

          if (cur[0] >= 1000) & (cur[1] >= 1000) & (cur[2] >= 1000):
            if (cur[2] == (cur[1] + 3330)) & (cur[1] == (cur[0] + 3330)):
              if is_prime(cur[0]) & is_prime(cur[1]) & is_prime(cur[2]):
                final_out.add((cur[0], cur[1], cur[2]))
  counter += 1

final_out


```
